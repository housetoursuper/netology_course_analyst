## Задание 1

### Исходные данные

````python
#Исходные данные:
# 1. Перечень всех документов

documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

# 2. перечень полок, на которых хранятся документы 
# (если документ есть в documents, то он обязательно должен быть и в directories)

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
````

#### Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру

Примеры работы:

1.

Введите команду:

Результат:

````python
#p Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру
def p():
    doc_number = input("Введите номер документа:")
    for doc in documents:
        if doc["number"] == doc_number:                     
            print('Владелец документа: ', doc["name"])
            break

    else:
        print("Документ не найден в базе")

p()
````

#### Пункт 4. Пользователь по команде "ads" может добавить новую полку

```
ads
```

```
1
```
Результат:  
`Такая полка уже существует. Текущий перечень полок: 1, 2, 3.`


````python
# Пункт 4. Пользователь по команде "ads" может добавить новую полку
def ads():
    shelf_id = input("Введите номер полки: ")
    for dir_ in directories.items():
        if shelf_id in dir_[0]:
            print(f"Такая полка уже существует. Текущий перечень полок: {', '.join(directories)}")
            break
    else:
        directories[shelf_id] = []
        print(f"Полка добавлена. Текущий перечень полок: {', '.join(directories)}")
ads()
````

````python
# Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)

def ds():
    shelf_id = input("Введите номер полки: ")
    if shelf_id in directories:
        if len(directories[shelf_id]) > 0:
            print(f"На полке есть документs, удалите их перед удалением полки. Текущий перечень полок: {', '.join(directories)}")
        else:
            del directories[shelf_id]
            print('Полка удалена. Текущий перечень полок: ' + ', '.join(directories))
    else:
        print ('Такой полки не существует. Текущий перечень полок:' + ', '.join(directories))

ds()   
````

## Основное меню запросов

````python
def main():
    while True:
        command = input('Введите команду - p, s, l, ads или ds, выход - q: ')
        if command == 'p':
            p()
        elif command == 's':
            s()
        elif command == 'l':
            l()
        elif command == 'ads':
            ads()
        elif command == 'ds':
            ds()
        elif command == 'q':
            print("До свидания")
            break
        else:
            print('Неккоректная команда! Введи новую команду')
            
main()
````
