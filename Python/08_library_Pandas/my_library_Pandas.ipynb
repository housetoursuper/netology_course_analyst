```python
# Загружаем по ссылке (https://files.grouplens.org/datasets/movielens/ml-latest-small.zip) датасеты ratings.csv и movies.csv на диск С:
ratings = pd.read_csv('C:\\Users\\Ryzen\\Desktop\\DataSet\\Data Analyst\\2_library_Pandas\\ml-latest-small\\ratings.csv')
movies = pd.read_csv('C:\\Users\\Ryzen\\Desktop\\DataSet\\Data Analyst\\2_library_Pandas\\ml-latest-small\\movies.csv')
```

```python
# сливаем датасеты ratings.csv и movies.csv в единый датасет data по ключу movieId 
data = ratings.merge(movies,on = 'movieId')
data.head()
```

```python
# формируем датасет df_5_title, состоящий из двух столбцов - rating (рейтинг: оценка = 5.0) и title (название фильма)
df_5_title = data[data['rating'] == 5.0]['title']
df_5_title.head()
```

```python
# Функция value_counts() используется для получения Series, содержащего уникальные значения. Она вернет результат, отсортированный в порядке убывания, так что первый элемент в коллекции будет самым встречаемым. NA-значения не включены в результат. Синтаксис df['your_column'].value_counts() — вернет количество уникальных совпадений в определенной колонке.
```

```python
# Формируем рейтинг ТОП-10 фильмов по количеству оценок 5.0
data[data['rating'] == 5.0]['title'].value_counts().head(10)
```

#### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
ds = pd.read_csv('C:\\Users\\Ryzen\\Desktop\\DataSet\\Data Analyst\\2_library_Pandas\\power.csv', 'r', delimiter=',')
ds.head()
```

```python
summ = 0

# формируем датасет ds_year_quantity по фильтру №1 - year:период: с 2005 по 2010 и исключить отрицательные значения quantity: 
ds_year_quantity = ds[(ds['year'] >= 2005) & (ds['year'] <= 2010) & (ds['quantity'] >= 0)]

# формируем датасет ds_category по фильтру №2 - category 4, 12 и 21 + условия фильтра №1:
ds_category = ds_year_quantity[ds_year_quantity.category.isin([4,12,21])]

# расчёт суммарного потребления стран Прибалтики (Латвия, Литва и Эстония) с учётом фильтров №1 и 2:
summ = ds_category[ds_category.country.str.contains('|'.join(['Lat','Lit','Est']))]['quantity'].sum()

print('суммарное потребление стран Прибалтики (Латвия, Литва и Эстония): {}'.format(summ))
```
