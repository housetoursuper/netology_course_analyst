Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш.

777: 200 руб.
999: 100 руб.
555: 50 руб.
333: 15 руб.
111: 10 руб.
X77: 5 руб.
XX7: 3 руб.
X00: 2 руб.
XX0: 1 руб.

Вопросы:

Выгодна ли игра игроку?
Сколько в среднем приобретает или теряет игрок за одну игру?
**Дополнительное задание повышенной сложности. **

Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования

– любая цифра

```python
import numpy as np
```

```python
trials = 1000  # Количество попыток
win = 0    # Количество успехов
sum_price_win = 0 # Сумма выигрыша
sum_bet = 1 # цена ставки 1 руб.

for i in range(trials):
    # Сотни:Десятки:Единицы: Случайное количество от 0 до 9
    roll_X00 = np.random.randint(0, 10)
    roll_0X0 = np.random.randint(0, 10)
    roll_00X = np.random.randint(0, 10)
    sum_bet = i + 1
    
    #777:200р
    if roll_X00 == 7 and roll_0X0 == 7 and roll_00X == 7:
        win = win + 1
        sum_price_win = sum_price_win + 200
        print (sum_bet, "|", roll_X00, roll_0X0, roll_00X)
    #999:100р
    if roll_X00 == 9 and roll_0X0 == 9 and roll_00X == 9:
        win = win + 1
        sum_price_win = sum_price_win + 100
        print (sum_bet, "|", roll_X00, roll_0X0, roll_00X)
    #555:50р
    if roll_X00 == 5 and roll_0X0 == 5 and roll_00X == 5:
        win = win + 1
        sum_price_win = sum_price_win + 50
        print (sum_bet, "|", roll_X00, roll_0X0, roll_00X)
    #333:15р
    if roll_X00 == 3 and roll_0X0 == 3 and roll_00X == 3:
        win = win + 1
        sum_price_win = sum_price_win + 15
        print (sum_bet, "|", roll_X00, roll_0X0, roll_00X)
    #111:10р
    if roll_X00 == 1 and roll_0X0 == 1 and roll_00X == 1:
        win = win + 1
        sum_price_win = sum_price_win + 10
        print (sum_bet, "|", roll_X00, roll_0X0, roll_00X)
    #Х77:5р
    if roll_X00 != 7 and roll_0X0 == 7 and roll_00X == 7:
        win = win + 1
        sum_price_win = sum_price_win + 5
        print (sum_bet, "|", roll_X00, roll_0X0, roll_00X)
    #ХХ7:3р
    if roll_X00 != 7 and roll_0X0 != 7 and roll_00X == 7:
        win = win + 1
        sum_price_win = sum_price_win + 3
        print (sum_bet, "|", roll_X00, roll_0X0, roll_00X)
    #Х00:2р
    if roll_0X0 == 0 and roll_00X == 0:
        win = win + 1
        sum_price_win = sum_price_win + 2
        print (sum_bet, "|", roll_X00, roll_0X0, roll_00X)
    #ХХ0:1р
    if roll_0X0 != 0 and roll_00X == 0:
        win = win + 1
        sum_price_win = sum_price_win + 1
        print (sum_bet, "|", roll_X00, roll_0X0, roll_00X)
```

```python
print("БАЗОВАЯ СТАТИСТИКА ИГР:")
print("Всего сделано ставок =", trials, ", из них: побед = ", win)
print("Сумма выигрыша, руб. =", sum_price_win)
print("Сумма ставок, руб. =", 0 - sum_bet)
score = sum_price_win - sum_bet
print("Итого на счету, руб. =", score)
p_monte_carlo = win / trials    # Вероятность из моделирования
print("вероятность выигрыша = ", p_monte_carlo)
```

```python
# Расчетная вероятность выигрыша:
# 777: 200 руб. 1/10*1/10*1/10 = 0,001
# 999: 100 руб. 1/10*1/10*1/10 = 0,001
# 555: 50 руб. 1/10*1/10*1/10 = 0,001
# 333: 15 руб. 1/10*1/10*1/10 = 0,001
# 111: 10 руб. 1/10*1/10*1/10 = 0,001
# *77: 5 руб. 1*1/10*1/10 = 0,01
# **7: 3 руб. 1*1*1/10 = 0,1
# *00: 2 руб. 1*1/10*1/10 = 0,01
# **0: 1 руб. 1*1*1/10 = 0,1

# итоговая расчетная вероятность:
# 5*0,001+2*0,01+2*0,1 = 0,225
```

```python
print("ОТВЕТЫ НА ВОПРОСЫ:")
print("--------------------")
if score > 0:
    print("Выгодна ли игра игроку?: игра выгодна игроку!")
else:
    print("Выгодна ли игра игроку?: игра НЕ выгодна игроку!")
print("--------------------")
print("Сколько в среднем приобретает или теряет игрок за одну игру?:")
print("Игрок теряет, руб = ", sum_bet/trials)
print("Игрок приобретает, руб = ", sum_price_win/trials)
print("Средний выигрыш за победу, руб = ", sum_price_win/win)
```
